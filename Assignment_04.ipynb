Q4 — Count Frequency of Numbers
numbers = [1,2,2,3,3,3,4,4,4,4]
Create frequency dictionary.

```python
#Solution
numbers = [1,2,2,3,3,3,4,4,4,4]

freq = {}

for num in numbers:
    if num in freq:
        freq[num] += 1

    else:
       freq[num] = 1

print(freq)
print(f"Frequency of Number: {len(freq)}")            
    
```

Q5 — Word Counter
Take sentence input and count word frequency

```python
#Solution
sentence = input("Enter a sentence: ").lower()
words = sentence.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1

    else:
        word_count[word] = 1
        
print(word_count)
print(f"Frequency of word count is : {len(words)}")             
```

Q6 — Student Marks & Average
marks = {"Math":80, "English":70, "Physics":90}
Calculate total and average.

```python
#Solution
Marks = {
    "Math": 80,
    "English": 70,
    "Physics": 90
}
total_marks = sum(Marks.values())
print(f"Total marks is: {total_marks}")

average = (total_marks) / len(Marks)
print(f"Average is: {average}")
```

Q7 — Highest Scoring Student
students = {
"Ali":85,
"Sara":92,
"Ahmed":78,
"Zara":95
}
Find highest scoring student.


```python
#Solution
Students = {
    "Ali": 85,
    "Sara": 92,
    "Ahmad": 78,
    "Zara": 95
}
highest_score = 0
top_student = ""

for key, score in Students.items():
    if score > highest_score:
        highest_score = score
        top_student = key

print(f"Highest scoring students is: {top_student} : {highest_score}")            
```
