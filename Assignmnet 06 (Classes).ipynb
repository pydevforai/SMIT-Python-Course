1. Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.

```python
# Add your code here
class Circle():
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.142 * (self.radius ** 2)

    def perimeter(self):
        return 2 * 3.142 * self.radius
        
r = float(input('Enter a radius: ')) 

c = Circle(r)
print(round(c.area(), 2))
print(round(c.perimeter(), 2))        
```

2. Write a Python program to create a person class. Include attributes like name, country and date of birth. Implement a method to determine the person's age.


```python
# Add your code here
class Person():
    def __init__(self, name, country, DOB):
        self.name = name 
        self.country = country
        self.DOB = DOB

    def find_age(self, cur_year):
        self.cur_year = cur_year
        parts = self.DOB.split('/')
        day = parts[0]
        month = parts[1]
        year = int(parts[2])
        age = self.cur_year - year
        print(f'Name: {self.name}')
        print(f'Country: {self.country}')
        print(f'Date of birth: {day}/{month}/{year}')
        print(f'Age: {age}')

obj = Person("Simra","Pakistan","15/12/2005")
obj.find_age(2025)        
           
```

3. Write a Python program to create a calculator class. Include methods for basic arithmetic operations.

```python
# Add your code here
class Calculator():
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2   

    def sub(self):
        return self.num1 - self.num2   

    def mul(self):
        return self.num1 * self.num2   

    def div(self):
        return self.num1 / self.num2  

    def mod(self):
        return self.num1 % self.num2 

    def power(self):
        return self.num1 ** self.num2
    
    def floordiv(self):
        return self.num1 // self.num2

obj = Calculator(4,2)
print(obj.add())
print(obj.sub())  
print(obj.mul())
print(obj.div())
print(obj.mod())
print(obj.power())
print(obj.floordiv())
```
